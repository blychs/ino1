*Importo los paquetes que necesito.* Esto depende del lenguaje que usen.

```python
import numpy as np
import matplotlib.pyplot as plt
```

Defino la derivada $\frac{d[NO_2]}{dt} = -v$.

```python
def derivada_no2(NO2):
    return -((NO2 ** 2.) * 0.54)
```

Calculo la concentración que me interesa usando que $f(t) \approx f(t_0) + \frac{df(t_0)}{dt}(t-t_0)$.

```python
t = np.arange(0, 500, 0.1)
NO2_0 = 0.021
NO2 = list([NO2_0])
NO = list([0])
O2 = list([0])
for i in range(0,len(t) -1 ):
    NO2.append(NO2[i] + derivada_no2(NO2[i]) * (t[i+1] - t[i]))
    NO.append(NO[i] - derivada_no2(NO2[i]) * (t[i+1] - t[i]))
    O2.append(O2[i] - 0.5 * derivada_no2(NO2[i]) * (t[i+1] - t[i]))
```

Por último, grafico todo

```python
plt.plot(t, NO2, label='NO$_2$');
plt.plot(t, NO, label='NO');
plt.plot(t, O2, label='O$_2$');
plt.xlabel('t(s)');
plt.ylabel('Concentración (M)');
plt.legend(loc=1);
```
